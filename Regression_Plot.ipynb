```python
import matplotlib.pyplot as plt
import seaborn as sns
```

A regression plot (often created using libraries like Seaborn in Python, as you've imported) is a visual tool that helps you understand the relationship between two variables. Imagine you have a scatter plot showing many data points, where each point represents a pair of values for two different things (e.g., a person's height and their weight).

A regression plot adds a 'regression line' to this scatter plot. This line is essentially the 'best fit' line that goes through the data points, showing the general trend. It helps you see if, for example, as one variable increases, the other also tends to increase (a positive relationship), or decrease (a negative relationship), or if there's no clear relationship

# lmplot()

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, hue= 'sex', palette= 'coolwarm')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
    scatter_kws={'s':300}) #scatter_kws markers ka size adjust kr deta h
```

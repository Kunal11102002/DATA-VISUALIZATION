```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:
distplot/histplot



*   displtot/histplot
*   jointplot


*   pairplot
*   rugplot














```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'], bins = 100 , kde = True) #bins ki value inc/dec kr skte h. histogram me jo curve h usme kernel density estimator banta hai ek curve jaisa. skewness ka pta chlta h.
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

```python
sns.pairplot(df) #scatter plot variance ya corelation find out krne k liye use hota h
```

```python
sns.pairplot(df,hue='size',palette='rainbow') #hue jo ha category wise divide kr deta h jaisa male ne zyada tip di h niche kyuki blue dots zayda h upr bhi h
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

A rug plot is a very simple plot that shows the distribution of a single variable. Imagine a number line, and for every data point, you draw a small vertical line (like a 'rug' fiber) on that number line. Where the lines are denser, there are more data points, indicating a higher frequency or concentration of values. It's often used with other plots like histograms or distribution plots to add more detail about individual data points

```python
sns.rugplot(df['tip'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill' , 'tip','size']]
tipscorr
```

```python
tipscorr.corr()  #corr: corelations
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

# Cluster map

```python
sns.clustermap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

# pivot table Heat map

```python
pvtflight = flights.pivot_table(index='month', columns='year', values='passengers')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
